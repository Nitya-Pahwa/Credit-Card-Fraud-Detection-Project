Dealing with Imbalanced Data

Using Under Sampling 

Since the data set of normal transaction is very high so we can reduce the majority class to count of minority class by under sampling

Splitting the data using imblearn module

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus=RandomUnderSampler(random_state=40)
```

```python
X_re,Y_re=rus.fit_resample(X,Y)
```

```python
X_re
```

```python
Y_re
```

```python
Y_re.value_counts()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_re,Y_re,test_size=0.2,random_state=42)
```

Reducing normal to 492 and then making a new dataset

```python
normal_sample=normal.sample(n=492)
```

```python
new=pd.concat([normal_sample,fraud],axis=0)
```

```python
new .describe()
```

```python
new.shape
```

```python
new['Class'].value_counts()
```

```python
new.groupby('Class').mean()
```

```python
new_X=new.drop(columns=['Class'])
```

```python
new_Y=new['Class']
```

```python
X_ntr,X_nte,Y_ntr,Y_nte=train_test_split(new_X,new_Y,test_size=0.2,random_state=42)
```

Applying Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lrn=LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lrn.fit(X_ntr,Y_ntr)
```

Calculating Accuracy Score

```python
from sklearn.metrics import accuracy_score
```

```python
Y_pred=lr.predict(X_test)
Y_npr=lrn.predict(X_nte)
```

```python
acc=accuracy_score(Y_test,Y_pred)
accn=accuracy_score(Y_nte,Y_npr)
```
